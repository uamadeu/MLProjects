- Once x, y and z variable have correlation greater than 0.9 between themselves, might be better the change them to a new variable called volume. For that we could approximate the diamond volume by a pyramid.

```python
diamond_df['delta'] = diamond_df['x'] - diamond_df['y']
diamond_df['delta'].describe()
diamond_df.loc[diamond_df['delta']==-50.81, :]
```

```python
diamond_df.iloc[11182, :]
```

```python
diamond_df.drop(11182, axis='rows', inplace=True)
```

Based on the size of the sample and the analysis of delta = x-y value, we can approximate the base area of the diamond by a hexagon inscribed within a square of side x.

```python
def hexagonside(a):
 
    # Side cannot be negative
    if (a < 0):
        return -1;
 
    # Side of the hexagon
    x = 0.5176 * a;
    return x;
```

```python
diamond_df['area'] = diamond_df['x'].apply(lambda x: hexagonside(x))
diamond_df
```

```python
diamond_df['volume'] = 1/3*diamond_df['area']*diamond_df['z']
diamond_df
```

Now that we have the volume variable calculated from the dimensions x, y and z, we should drop all the redundant variables.

```python
df2 = diamond_df.drop(['x', 'y', 'z', 'delta', 'area'], axis='columns')
df2
```

# 3. Machine Learning

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, mean_absolute_error, mean_squared_error, r2_score
```

- Since table and depth variables doesn't have a good correlation between price, might be better to simply drop them from our model.

```python
x = df3.drop(['price', 'table', 'depth'], axis=1)
y = df3.price
x
```

## 3.1 Dummies

```python
x_dummy = pd.get_dummies(x, columns=['cut', 'color', 'clarity'])
x_dummy
```

```python
y
```

```python
categorical = ['cut', 'color', 'clarity']
numerical = ['carat', 'depth', 'table', 'volume']
```

## 3.2 Loading models and pipelines

```python
ada = AdaBoostRegressor()
rf = RandomForestRegressor()
lr = LinearRegression()

pipe1 = make_pipeline(StandardScaler(), ada)
pipe2 = make_pipeline(StandardScaler(), rf)
pipe3 = make_pipeline(StandardScaler(), lr)
```

## 3.3 Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x_dummy, y, test_size=0.3, random_state=42)
```

```python
pipe1.fit(x_train, y_train)
y_pred = pipe1.predict(x_test)
kfold = KFold(n_splits=10)
s = cross_val_score(ada, x_train, y_train, cv=kfold)
s.mean()
```

```python
r2_score(y_test, y_pred)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'mse: {mse}, mae: {mae}')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred), max(y_test))
p2 = min(min(y_pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
pipe2.fit(x_train, y_train)
y_pred2 = pipe2.predict(x_test)
kfold = KFold(n_splits=10)
s = cross_val_score(rf, x_train, y_train, cv=kfold)
s.mean()
```

```python
r2_score(y_test, y_pred2)
```

```python
mse = mean_squared_error(y_test, y_pred2)
mae = mean_absolute_error(y_test, y_pred2)
print(f'mse: {mse}, mae: {mae}')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred2, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred2), max(y_test))
p2 = min(min(y_pred2), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
pipe3.fit(x_train, y_train)
y_pred3 = pipe3.predict(x_test)
kfold = KFold(n_splits=10)
s = cross_val_score(lr, x_train, y_train, cv=kfold)
s.mean()
```

```python
r2_score(y_test, y_pred3)
```

```python
mse = mean_squared_error(y_test, y_pred3)
mae = mean_absolute_error(y_test, y_pred3)
print(f'mse: {mse}, mae: {mae}')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred3, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred3), max(y_test))
p2 = min(min(y_pred3), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

- Although, both Mean Squared Error (MSE) and Mean Absolute Error (MAE) for testing samples are higher than expected. This means the model might have more outliers than expected.
- The best model is the Random Forest Regressor with 98% accuracy during training.
- Random Forest showed the significant lower values for MSE and MAE when compared to the other models.
- After add topic 2.2 and remove some outliers from the analysis, the values of MSE and MAE for the three models reduced significantly. For instance, the AdaBoost was getting 1900000 MSE value before removing outliers and got 894808 MSE after.
