```python
eco = []
name = []
for i in range(len(detailed_base.opening)):
    if type(detailed_base.opening[i]) is dict:
        eco.append(detailed_base.opening[i]['eco'])
        name.append(detailed_base.opening[i]['name'])
    else:
        eco.append('None')
        name.append('None')
```

```python
player_info = pd.DataFrame(list(detailed_base['players']))
opening = pd.DataFrame(data = {'ECO': eco, 'opening_name': name})
opening
```

```python
complete_df = pd.concat([detailed_base, player_info, opening], axis = 'columns')
complete_df.drop(['id', 'lastMoveAt', 'players', 'opening'], axis = 'columns', inplace = True)
```

```python
white_data = []
black_data = []
errors = []
for index in range(len(complete_df)):
    try:
        white_data.append([str(complete_df.white[index]).split(': ')[3], str(complete_df.white[index]).split(': ')[4][0:4]])
        black_data.append([str(complete_df.black[index]).split(': ')[3], str(complete_df.black[index]).split(': ')[4][0:4]])
    except IndexError:
        print(f'{index} IndexError, white-black: {len(white_data) - len(black_data)}')
        errors.append(index)
```

```python
complete_df.drop(complete_df.index[errors], inplace=True)
complete_df.reset_index(inplace=True)
len(complete_df)
```

```python
white_data= []
black_data = []
for index in range(len(complete_df)):
    try:
        white_data.append([str(complete_df.white[index]).split(': ')[3], str(complete_df.white[index]).split(': ')[4][0:4]])
        black_data.append([str(complete_df.black[index]).split(': ')[3], str(complete_df.black[index]).split(': ')[4][0:4]])
    except KeyError:
        print(f'{index} KeyError, white-black: {len(white_data) - len(black_data)}')
```

```python
white_player_refined = [[data[0].split('},')[0], data[1]] for data in white_data]
black_player_refined = [[data[0].split('},')[0], data[1]] for data in black_data]
```

```python
white = pd.DataFrame(white_player_refined, columns=['white_player', 'rating'])
black = pd.DataFrame(black_player_refined, columns=['black_player', 'rating'])
```

```python
df2 = pd.concat([complete_df, white, black], axis = 'columns')
df2.drop(['index', 'white', 'black'], axis='columns', inplace=True)
df2
```
