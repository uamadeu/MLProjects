# Treating data

### Droping unnecessary columns

```python
df_train.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis='columns', inplace=True)
df_test.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis='columns', inplace=True)
```

```python
df_train = pd.get_dummies(df_train, columns=['Sex', 'Embarked'], drop_first=True)
df_test = pd.get_dummies(df_test, columns=['Sex', 'Embarked'], drop_first=True)
df_train
```

### Imputing missing data

```python
imputer = SimpleImputer(strategy='median')
imputed_train = imputer.fit_transform(df_train)
imputed_test = imputer.fit_transform(df_test)
```

```python
sca = StandardScaler()
lenc = LabelEncoder()
X_train = sca.fit_transform(imputed_train)
X_train = pd.DataFrame(X_train)
X_test = sca.fit_transform(imputed_test)
X_test = pd.DataFrame(X_test)
```

```python
Y_train = X_train[0]
X_train.drop([0], axis='columns', inplace=True)
X_train
```

### Label Encoding Y_train

```python
Y_train = lenc.fit_transform(Y_train)
```

### Loading the models e calculating AUC

```python
predictions = {}
for model in [LogisticRegression(), 
              DecisionTreeClassifier(), 
              KNeighborsClassifier(), 
              GaussianNB(), 
              SVC(probability=True), 
              RandomForestClassifier()
             ]:
    
    cls = model
    cls.fit(X_train, Y_train)
    pred = cls.predict(X_test)
    predictions[cls] = pred
    kfold = model_selection.KFold(n_splits=10)
    s = model_selection.cross_val_score(cls, X_train, Y_train, scoring='roc_auc', cv=kfold)
        
    print(f"{model} AUC:"
          f"{s.mean():.3f} STD: {s.std():.2f}")
    
    roc = yellowbrick.ROCAUC(cls)
    roc.fit(X_train, Y_train)
    roc.score(X_train, Y_train)
    roc.show()
```

```python
predict = pd.DataFrame(data={list(predictions.keys())[0]:list(predictions.values())[0], 
                             'DecisionTreeClassifier()':list(predictions.values())[1],
                             list(predictions.keys())[2]:list(predictions.values())[2],
                             list(predictions.keys())[3]:list(predictions.values())[3],
                             list(predictions.keys())[4]:list(predictions.values())[4],
                             'RandomForestClassifier()':list(predictions.values())[5]
                            })
```
